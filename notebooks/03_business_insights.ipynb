# Business Insights 

## Contexto
Este proyecto analiza el comportamiento de compras, performance operativa y patrones de consumo dentro de una plataforma de ecommerce utilizando datos transaccionales reales.

Se desarrolló un pipeline completo que incluye:
- Procesamiento y modelado de datos mediante SQL
- Análisis exploratorio (EDA)
- Análisis analítico orientado a negocio (Analytics)

El objetivo de este notebook es traducir los hallazgos técnicos en insights accionables para la toma de decisiones de negocio.

## Insight 1 – Concentración geográfica del revenue

El análisis por estado muestra una fuerte concentración del revenue en regiones específicas, destacándose principalmente São Paulo como el principal generador de ingresos.

Sin embargo, el ticket promedio más alto no se encuentra necesariamente en los estados con mayor revenue total, lo que sugiere diferencias entre volumen de compras y gasto individual.

Además, se observa que los estados con mayor revenue tienden a presentar menores tiempos de entrega, posiblemente debido a mayor infraestructura logística.


- Potencial de expansión en estados con alto ticket promedio pero bajo volumen
- Importancia de la infraestructura logística en regiones de alto revenue
- Necesidad de estrategias regionales diferenciadas

## Insight 2 – Segmentación de clientes por nivel de gasto

Los clientes fueron segmentados en tres grupos según su nivel de gasto: low, medium y high.

Se observa que el segmento high concentra una proporción significativa del revenue total, mientras que los segmentos low y medium aportan mayor volumen de órdenes.


- Estrategias de fidelización para clientes de alto gasto
- Promociones para incrementar el gasto en segmentos medios
- Estrategias de volumen para segmentos de bajo gasto

## Insight 3 – Patrones temporales de compra

El análisis temporal muestra que las compras se concentran principalmente entre las 10 y las 21 horas, con picos cercanos al mediodía.

Los días de semana presentan mayor volumen de compras en comparación con los fines de semana.

- Optimización de campañas publicitarias en horarios pico
- Refuerzo de soporte operativo en momentos de mayor demanda
- Diseño de promociones para horarios de baja actividad

## Insight 4 – Preferencias en métodos de pago

Se observa una fuerte concentración en determinados métodos de pago, lo que permite identificar preferencias claras de los usuarios al momento de completar sus compras.

- Optimización de la experiencia de usuario en los métodos más utilizados
- Desarrollo de promociones específicas por medio de pago
- Evaluación de planes de cuotas y financiamiento

## Resumen ejecutivo

| Área | Insight | Acción sugerida |
|------|--------|----------------|
| Geografía | Revenue concentrado | Estrategias regionales |
| Clientes | High spenders dominan revenue | Programas de fidelización |
| Operaciones | Horarios pico definidos | Optimización operativa |
| Pagos | Métodos dominantes | Mejora UX y promociones |
