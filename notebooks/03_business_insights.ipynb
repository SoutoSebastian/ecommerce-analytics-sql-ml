# Business Insights 

## Context
This project analyzes purchasing behavior, operational performance, and consumption patterns within an e-commerce platform using real transactional data.

A complete pipeline was developed, including:
- Data processing and modeling using SQL
- Exploratory Data Analysis (EDA)
- Business-oriented analytical analysis (Analytics)

The goal of this notebook is to translate technical findings into actionable insights for business decision-making.

## Insight 1 – Geographic Concentration of Revenue

The state-level analysis reveals a strong concentration of revenue in specific regions, with São Paulo standing out as the primary income generator.

However, the highest average order value is not necessarily found in the states with the highest total revenue, suggesting a distinction between purchase volume and individual spending.

Furthermore, states with higher revenue tend to have shorter delivery times, likely due to better logistics infrastructure.

- Expansion potential in states with high average order value but low volume
- Importance of logistics infrastructure in high-revenue regions
- Need for differentiated regional strategies

## Insight 2 – Customer Segmentation by Spending Level

Customers were segmented into three groups based on their spending level: low, medium, and high.

It is observed that the high segment accounts for a significant proportion of total revenue, while the low and medium segments contribute a higher volume of orders.


- Loyalty strategies for high-spending customers
- Promotions to increase spending in medium segments
- Volume strategies for low-spending segments

## Insight 3 – Temporal Purchase Patterns

The temporal analysis shows that purchases are primarily concentrated between 10:00 and 21:00 (10 AM and 9 PM), with peaks around midday.

Weekdays show a higher volume of purchases compared to weekends.

- Optimize ad campaigns for peak hours
- Strengthen operational support during high-demand periods
- Design promotions for low-activity hours

## Insight 4 – Payment Method Preferences

A strong concentration in specific payment methods is observed, allowing for the identification of clear user preferences at checkout.

- Optimize the user experience for the most-used payment methods
- Develop specific promotions tied to payment methods
- Evaluate installment and financing plans

## Resumen ejecutivo

| Area | Insight | Suggested Action |
|------|--------|----------------|
| Geography | Revenue is highly concentrated in specific states | Develop targeted regional strategies |
| Customers | High-spending customers dominate total revenue | Implement loyalty programs for high-value segments |
| Operations | Clear peak hours identified | Optimize operational staffing and campaigns for peak times |
| Payments | Dominant payment methods identified | Improve user experience and create tailored promotions for preferred methods |
